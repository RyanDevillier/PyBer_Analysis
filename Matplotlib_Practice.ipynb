```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Creating the previous plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Creating the previous plot with ax.plt() but unpacked
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Creating the previous plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Skill Drill 5.1.7
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]



fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color = "skyblue", 
           alpha = 0.8, edgecolor = "black", lw = 2, label = "Boston")

ax.set_xlim(0,50)

ax.invert_yaxis()
ax.legend()

ax.set_title("PyBer Fare by Month")
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Increases figure size
plt.subplots(figsize = (8, 8))

# Array that contains the colors for each wedge of the pie chart
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Array for the values we want to explode that correspond to the values in x_axis
explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)

plt.pie(y_axis, 
        explode = explode_values, 
        labels = x_axis, 
        autopct = '%.1f%%', 
        colors = colors)

plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


fig, ax = plt.subplots()

plt.gcf().set_size_inches(8, 8)

explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis, labels = x_axis, 
       autopct = '%.1f%%', 
       explode = explode, 
       shadow = True, 
       startangle = 95,
       counterclock = False,
       colors = colors)

plt.show()
```
